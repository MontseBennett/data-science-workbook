# BeautifulSoup Tutorial

```
import requests
import csv
from bs4 import BeautifulSoup
```

```
src = requests.get('https://alphamartian.wordpress.com/', 'lxml')
```

```
print(src)
```

```
soup = BeautifulSoup(src.text)
print(soup.prettify())
```

## Get the first article

```
print(soup.find('article').prettify())
```

```
title = soup.find('article').find('a').text
print(title)
```

```
link = soup.find('article').find('a')['href']
print(link)
```

```
text_content = soup.find('article').find('div', class_='entry-content').p.text
print(text_content)
```

## Get all the articles

```
csv_file = open('scrapped.csv', 'w')
```

```
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['title', 'link', 'text_content'])
```

```
for article in soup.find_all('article'):
    try:
        title = article.find('a').text
        link = article.find('a')['href']
        text_content = article.find('div', class_='entry-content').p.text
        print(title)
        print(link)
        print(text_content)
        print('\n')
        csv_writer.writerow([title, link, text_content])
    except:
        continue
```

```
csv_file.close()
```
